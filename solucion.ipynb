## FASE 2: LIMPIEZA Y TRANSFORMACIÓN DEL DATASET

A continuación se aplican transformaciones sistemáticas a cada problema identificado en Fase 1.

```python
# 2.0 Preparación: Crear copia para limpieza
df_clean = df.copy()
```

```python
# 2.1 Limpieza de Texto: Names (Columna Name)
print("- Convertir a Title Case para consistencia")
print("- Remover títulos variados (Mr., Ms., Dr., Mrs., mRS., MS.)")
print("- Limpiar espacios extra\n")

def clean_name(name):
    """Limpia nombres: título case, sin títulos, sin espacios extras"""
    # Remover títulos comunes
    titulos = ['Mr.', 'Ms.', 'Mrs.', 'Dr.', 'mRS.', 'MS.', 'mr.', 'mrs.', 'dr.', 'MD', 'Sr.']
    cleaned = name.strip()
    
    for titulo in titulos:
        cleaned = cleaned.replace(titulo, '').strip()
    
    # Title case
    cleaned = ' '.join([word.capitalize() for word in cleaned.split()])
    
    return cleaned

print("ANTES (primeros 10):")
print(df_clean['Name'].head(10).tolist())

df_clean['Name'] = df_clean['Name'].apply(clean_name)

print("\nDESPUÉS (primeros 10):")
print(df_clean['Name'].head(10).tolist())
```

```python
# 2.2 Limpieza de Columna: Doctor
print("- Estandarizar nombre de doctores")
print("- Aplicar Title Case\n")

def clean_doctor(doctor):
    """Limpia nombres de doctores"""
    # Títulos y sufijos a remover
    sufijos = [' MD', ' PhD', 'Sr.', 'Dr. ', 'Mr.', 'Ms', 'Mrs. ', 'Miss ',' DDS', ' DO', ' DVM', ' MBA', 'MSc', 'RN', 'NP', 'PA', 'FACS']
    cleaned = doctor.strip()
    
    for sufijo in sufijos:
        cleaned = cleaned.replace(sufijo, '').strip()
    
    # Title case
    cleaned = ' '.join([word.capitalize() for word in cleaned.split()])
    
    return cleaned

print("ANTES (primeros 10):")
print(df_clean['Doctor'].head(10).tolist())

df_clean['Doctor'] = df_clean['Doctor'].apply(clean_doctor)

print("\nDESPUÉS (primeros 10):")
print(df_clean['Doctor'].head(10).tolist())
```

```python
# Lista extendida de títulos y sufijos comunes
titulos_sufijos2 = [' MD', ' PhD', 'Sr.', 'Dr. ', 'Mr.', 'Ms', 'Mrs. ', 'Miss ',' DDS', ' DO', ' DVM', ' MBA', 'MSc', 'RN', 'NP', 'PA', 'FACS']

# Detectar doctores con títulos/sufijos
doctores_con_titulo2 = df_clean[df_clean['Doctor'].str.contains('|'.join(titulos_sufijos2), case=False, na=False)]['Doctor'].unique()

print("Doctores con títulos/sufijos:", doctores_con_titulo2)
print("Número de doctores con títulos/sufijos:", len(doctores_con_titulo2))
```

```python
# 2.3 Limpieza de Columna: Hospital
print("- Remover comas finales")
print("- Remover espacios inconsistentes")
print("- Estandarizar formato de nombre\n")

def clean_hospital(hospital):
    """Limpia nombres de hospitales"""
    cleaned = hospital.strip()
    
    # Remover comas finales
    if cleaned.endswith(','):
        cleaned = cleaned[:-1].strip()
    
    # Remover comas internas múltiples
    cleaned = cleaned.replace(',,', ',').replace(', ,', ',')
    
    # Title case para mejor legibilidad
    cleaned = ' '.join([word.capitalize() for word in cleaned.split()])
    
    return cleaned

print("ANTES (primeros 10):")
for h in df_clean['Hospital'].head(10):
    print(f"  '{h}'")

df_clean['Hospital'] = df_clean['Hospital'].apply(clean_hospital)

print("\nDESPUÉS (primeros 10):")
for h in df_clean['Hospital'].head(10):
    print(f"  '{h}'")
```

```python
# 2.4 Conversión de Fechas y Validación
print("- Convertir a tipo datetime para operaciones temporales")
print("- Validar relaciones lógicas")
print("- Calcular duración de internación como métrica nueva\n")

# Convertir a datetime
df_clean['Date of Admission'] = pd.to_datetime(df_clean['Date of Admission'])
df_clean['Discharge Date'] = pd.to_datetime(df_clean['Discharge Date'])

# Validar que Discharge > Admission
fechas_invalidas = df_clean[df_clean['Discharge Date'] < df_clean['Date of Admission']]
if len(fechas_invalidas) > 0:
    print(f"\n ALERTA: {len(fechas_invalidas)} registros con fecha de discharge anterior a admission")
    print("  Manteniendo estos registros para revisión posterior")
else:
    print("\n Todas las fechas de discharg son posteriores a admission")

# Calcular duración de internación (nuevas métrica)
df_clean['Hospitalization_Days'] = (
    df_clean['Discharge Date'] - df_clean['Date of Admission']
).dt.days

print(f"\n Nueva columna 'Hospitalization_Days' calculada")
print(df_clean[['Date of Admission', 'Discharge Date', 'Hospitalization_Days']].head(5).to_string())
```

### Se omite de momento por temas de decimales y dinero no se realiza redondeo, depende el fin con el que se quiera realizar el cálculo.

```python
# 2.6 Validación de Datos Categóricos
print("\n" + "=" * 80)
print("2.6 - VALIDACIÓN: Datos Categóricos")
print("=" * 80)

validaciones = {
    'Gender': ['Male', 'Female'],
    'Blood Type': ['A+', 'A-', 'B+', 'B-', 'AB+', 'AB-', 'O+', 'O-'],
    'Medical Condition': ['Asthma', 'Arthritis', 'Cancer', 'Obesity', 'Diabetes', 'Hypertension'],
    'Admission Type': ['Emergency', 'Urgent', 'Elective'],
    'Medication': ['Penicillin', 'Paracetamol', 'Aspirin', 'Ibuprofen', 'Lipitor'],
    'Test Results': ['Normal', 'Abnormal', 'Inconclusive']
}

print("\nValidación de valores únicos:")
for col, valores_validos in validaciones.items():
    valores_actuales = set(df_clean[col].unique())
    valores_invalidos = valores_actuales - set(valores_validos)
    
    print(f"\n  {col}:")
    print(f"    Valores válidos esperados: {valores_validos}")
    print(f"    Valores actuales: {sorted(valores_actuales)}")
    
    if valores_invalidos:
        print(f"     INVÁLIDOS ENCONTRADOS: {valores_invalidos}")
    else:
        print(f"     Todos los valores son válidos")

print("\n Validación completada")
```

```python
# 2.7 Eliminación de Duplicados
print("\n" + "=" * 80)
print("2.7 - LIMPIEZA: Eliminación de Duplicados")
print("=" * 80)

duplicados_before = df_clean.duplicated().sum()
print(f"\nDuplicados encontrados antes: {duplicados_before}")

if duplicados_before > 0:
    print(f"  Eliminando {duplicados_before} registros duplicados...")
    df_clean = df_clean.drop_duplicates(keep='first')
    print(f"   Duplicados eliminados")
else:
    print("   No hay duplicados que eliminar")

print(f"\nDataset limpio: {df_clean.shape[0]} filas x {df_clean.shape[1]} columnas")
```

```python
# 2.8 Validación Final y Comparativa
print("\n" + "=" * 80)
print("2.8 - VALIDACIÓN FINAL Y COMPARATIVA")
print("=" * 80)

print("\n--- COMPARATIVA GENERAL ---")
print(f"\nDataset Original:  {df.shape[0]} filas x {df.shape[1]} columnas")
print(f"Dataset Limpio:    {df_clean.shape[0]} filas x {df_clean.shape[1]} columnas")
print(f"\nNueva columna agregada: 'Hospitalization_Days'")

print("\n--- VALIDACIÓN DE NULOS ---")
print(f"\nValores nulos en dataset limpio:")
nulos_clean = df_clean.isnull().sum()
print(nulos_clean[nulos_clean > 0] if nulos_clean.sum() > 0 else " No hay valores nulos")

print("\n--- VALIDACIÓN DE DUPLICADOS ---")
print(f"Duplicados: {df_clean.duplicated().sum()}")
```
